```python
import pandas as pd
import numpy as np
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv("Downloads/us-education-datasets-unification-project/states_all.csv")
```

```python
df.head()
```

```python
data = [go.Bar(x =df['STATE'], y = df['STATE_REVENUE'])]
layout = go.Layout(title="State Revenue",
                xaxis=dict(title='State'),
                yaxis=dict(title='State Revenue'))
fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
